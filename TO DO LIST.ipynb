```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("To-Do List with Priorities")
root.geometry("500x450")
root.config(bg="#e6e6e6")

tasks = []

def add_task():
    task = task_entry.get().strip()
    priority = priority_var.get()
    if task == "":
        messagebox.showwarning("Warning", "Please enter a task!")
    else:
        full_task = (task, priority)
        tasks.append(full_task)
        sort_tasks()
        display_tasks()
        task_entry.delete(0, tk.END)


def delete_task():
    try:
        index = tasks_listbox.curselection()[0]
        tasks.pop(index)
        display_tasks()
    except:
        messagebox.showwarning("Warning", "Please select a task to delete!")


def clear_all():
    tasks.clear()
    display_tasks()

def sort_tasks():
    priority_order = {"High": 1, "Medium": 2, "Low": 3}
    tasks.sort(key=lambda x: priority_order[x[1]])


def display_tasks():
    tasks_listbox.delete(0, tk.END)
    for t, p in tasks:
        tasks_listbox.insert(tk.END, f"{t}  [{p}]")
task_label = tk.Label(root, text="Enter Task:", font=("Arial", 12), bg="#e6e6e6")
task_label.pack(pady=5)

task_entry = tk.Entry(root, font=("Arial", 12), width=30)
task_entry.pack(pady=5)

priority_label = tk.Label(root, text="Select Priority:", font=("Arial", 12), bg="#e6e6e6")
priority_label.pack(pady=5)

priority_var = tk.StringVar(value="Medium")
priority_menu = tk.OptionMenu(root, priority_var, "High", "Medium", "Low")
priority_menu.config(font=("Arial", 12))
priority_menu.pack(pady=5)

add_button = tk.Button(root, text="Add Task", command=add_task, bg="#4CAF50", fg="white", font=("Arial", 12))
add_button.pack(pady=10)

frame = tk.Frame(root)
frame.pack(pady=10, fill="both", expand=True)

tasks_listbox = tk.Listbox(frame, font=("Arial", 12), width=50, height=10)
tasks_listbox.pack(side="left", fill="both", expand=True)

scrollbar = tk.Scrollbar(frame, orient="vertical")
scrollbar.config(command=tasks_listbox.yview)
scrollbar.pack(side="right", fill="y")
tasks_listbox.config(yscrollcommand=scrollbar.set)

delete_button = tk.Button(root, text="Delete Task", command=delete_task, bg="#f44336", fg="white", font=("Arial", 12))
delete_button.pack(pady=5)

clear_button = tk.Button(root, text="Clear All", command=clear_all, bg="#555", fg="white", font=("Arial", 12))
clear_button.pack(pady=5)

root.mainloop()
```
