. Write code for A* Algorithm

```python
import heapq

class Node:
    def __init__(self, position, parent=None):
        self.position = position       # (row, col)
        self.parent = parent
        self.g = 0                     # cost from start
        self.h = 0                     # heuristic
        self.f = 0                     # g + h

    def __lt__(self, other):
        return self.f < other.f


def heuristic(a, b):
    # Manhattan Distance
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


def astar(grid, start, goal):
    open_list = []
    closed_list = set()

    start_node = Node(start)
    start_node.h = heuristic(start, goal)
    start_node.f = start_node.h

    heapq.heappush(open_list, start_node)

    while open_list:
        current = heapq.heappop(open_list)

        if current.position == goal:
            # Build final path
            path = []
            while current:
                path.append(current.position)
                current = current.parent
            return path[::-1]

        closed_list.add(current.position)

        # Possible moves (up, down, left, right)
        moves = [(0,1), (1,0), (0,-1), (-1,0)]

        for move in moves:
            new_pos = (current.position[0] + move[0],
                       current.position[1] + move[1])

            # Check boundaries
            if new_pos[0] < 0 or new_pos[0] >= len(grid) or \
               new_pos[1] < 0 or new_pos[1] >= len(grid[0]):
                continue

            # Check if blocked
            if grid[new_pos[0]][new_pos[1]] == 1:
                continue

            if new_pos in closed_list:
                continue

            new_node = Node(new_pos, current)
            new_node.g = current.g + 1
            new_node.h = heuristic(new_pos, goal)
            new_node.f = new_node.g + new_node.h

            heapq.heappush(open_list, new_node)

    return None


grid = [
    [0,0,0,0],
    [1,1,0,1],
    [0,0,0,0],
    [0,1,1,0]
]

start = (0,0)
goal = (3,3)

result = astar(grid, start, goal)

print("\nFinal Path:")
print(result)
```
